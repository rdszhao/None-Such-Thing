```python
# define a function that extracts all the artists from a playlist
from collections import Counter

# create a playlist class that makes things a little easier
class playlist:
    def __init__(self, playlist_id, head=client):
        self.obj = head.playlist(playlist_id=playlist_id)
        self.name = self.obj['name']
    
    def artists(self):
        tracks = self.obj['tracks']['items']
        artists = Counter()
        try:
            for track in tracks:
                artists.update(artist['name'] for artist in track['track']['artists'])
            return artists
        except:
            return Counter()
```

```python
mitre_10_id = '5EbNNdQvJ8U4TyuqetE5pY?si=wHISzC8XTBOIbsJ0oRB7yQ'
mitre_10 = playlist(mitre_10_id)
mitre_10.name, mitre_10.artists().most_common(10)
```

### 3. Just One Kevin

So we have a problem. Spotify doesn't explicitly link a user and his playlists, which means we'll have to manually get the playlists for each user by using a scraper.

From there, we aggregate all the artists he has on a playlist, for _every_ playlist.

We're going to use `Beautiful Soup`, a web scraping and html parsing library alongside a `Selenium` web driver. `Selenium` might feel like a bit of overkill, but it'll come in useful little later on.

```python
from time import sleep
from urllib.request import urlopen
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup as bs
```

```python
# set selenium options
options = Options()
options.add_argument('--window-size1920.1200')
options.headless = False
DRIVER_PATH = './chromedriver'
get_elements = "return document.getElementsByTagName('html')[0].innerHTML"

def __driver__(url, driver_core):
    'uses a url to start a selenium chromedriver'
    driver = driver_core
    if not driver:
        driver = webdriver.Chrome(options=options, executable_path=DRIVER_PATH)
    driver.get(url)
    return driver

class Driver:
    'wrapper for selenium chromedriver'
    def __init__(self, url, driver_core):
        'returns a driver using parameters passed to __driver__'
        self.core = __driver__(url, driver_core)
        sleep(5)
        self.source = self.core.execute_script(get_elements)

    def get(url):
        driver.get(url)

    def stop(self):
        self.core.quit()


class Soup:
    'interface for selenium scraper and beautiful soup'
    def __init__(self, src, driver_core=None, scroll=False, use_selenium=True):
        'returns bsoup object of driver source, given either url or d'
        if type(src) is str:
            if '.html' in src:
                self.base = bs(open(src))
            else:
                if not use_selenium:
                    self.base = bs(urlopen(src), features='lxml')
                else:
                    self.driver = Driver(src, driver_core)
                    self.core = self.driver.core
                    raw = self.driver.source
                    self.base = bs(raw, features='lxml')
        else:
            self.driver = src
            self.core = src
            raw = src.execute_script(get_elements)
            self.base = bs(raw, features='lxml')

    def extract(self, *tags, to_text=False, normalize=True, **attributes):
        'get tags and attributes from soup base with option for text conversion'
        findings = self.base(*tags, **attributes)
        if to_text:
            findings = [result.get_text() for result in findings]
            if normalize:
                findings = list(filter(None, [element.strip() for element in findings]))
        return findings
```

Since the API actually provides no information from the `profile` object, we only need the url for the profile, and not the id. From there, we scrape the urls of all his playlists from the source code.

```python
# scrolling output definition
from IPython.core.display import HTML
style = """
<style>
div.output_area {
    overflow-y: scroll;
}
div.output_area img {
    max-width: unset;
}
</style>
"""

nguyen = 'https://open.spotify.com/user/kevin.nguyen9852'
obj = Soup(nguyen) 
core = obj.core
```

Thankfully, the playlists are actually pretty easy to find, since they're all tagged with `<a>`:

```python
findings = obj.extract('a', to_text=True)
findings[9:19]
```

So we can get the names. That's cool. But how about the playlist IDs? Since those are what we need. On close inspection, they're the `href` attribute of the `a` class.

```python
findings = obj.extract('a')
for element in findings[9:19]:
    print(element['href'])
```

Oh yeah. It's all coming together.

```python
# a wrapper to make things easier
class Kevin:
    def __init__(self, url, core=None):
        try:
            self.soup = Soup(url, driver_core=core, use_selenium=False)
        except:
            self.soup = None

    def get_playlists(self):
        findings = self.soup.extract('a')
        self.playlist_ids = [element['href'].replace('/playlist/', '') for element in findings if 'playlist' in element['href']]
        self.playlists = [playlist(id) for id in self.playlist_ids]

    def artist_stats(self):
        try:
            self.get_playlists()
            return sum([playlist.artists() for playlist in self.playlists], Counter())
        except:
            return Counter()
```

```python
kevin = Kevin(nguyen)
kevin.get_playlists()

# list the 10 most frequent artists in his first playlist
kevin.playlists[0].artists().most_common(10)
```

There we go! Armed with only a user's url, we can now find and break down all of their playlists, and find out the artist density of their playlists:

```python
# make a quick function that makes the full aggregate of artists
artists = kevin.artist_stats()
artists.most_common(10)
```

Wow there are 809 _Kevin Nguyens_ on Spotify. We won't have any shortage of users then.

```python
kevins = [Kevin(user) for user in users]
stats = [kevin.artist_stats() for kevin in kevins if kevin]
overall = sum(stats, Counter())
```

```python
del overall['']
overall.most_common(20)
```

Wow! The 30 minutes it took this code to run was well worth it. Now what about proportions? Drake appeared 5768 times, but if I threw a dart at all these Kevins, what is the chance the my dart lands on a Kevin that listens to Drake?

```python
top_100 = overall.most_common(100)
proportions = {}
for artist, num in top_100:
    freq = 0
    for stat in stats:
        if artist in stat:
            freq += 1
    prop = freq / len(stats)
    proportions.update({artist: prop})
```

```python
sorted_proportions = sorted(proportions.items(), key=lambda x: x[1], reverse=True)
sorted_proportions[:10]
```

Basically half!

### 5. Public Enemy Numba One

So now that we know the artist distribution across _real_ Kevin Nguyens, how does that stack up against what we _perceive_ and _associate_ with him? We're going to use the `search` function of `spotipy`:

```python
def search_playlists(query, client):
    results = client.search(q=query, limit=50, type='playlist')
    playlists = results['playlists']['items']
    while results['playlists']['next']:
        results = client.next(results['playlists'])
        playlists.extend(results['playlists']['items'])
    return playlists
```

```python
results = search_playlists('kevin nguyen', client)
playlists = [playlist(result['id']) for result in results]
artists = sum([pls.artists() for pls in playlists], Counter())
```

```python
artists.most_common(59)
```

Wow. That's quite the difference! Guess everyone's idea of Kevin Nguyen is a lot more cultured than what he actually is.

```python
p_top_100 = artists.most_common(100)
p_top_200 = artists.most_common(200)
k_top_100 = [artist for artist, num in top_100]
k_top_200 = [artist for artist, num in overall.most_common(200)]
```

Let's do some chart testing.

```python
def artist_props(dist1, dist2):
    total = 0
    for artist, num in dist1:
        if artist in dist2:
            total += 1
    return total / len(dist1)

artist_props(p_top_100, k_top_100), artist_props(p_top_200, k_top_200)
```

This is really cool. While the users are pretty bad at predicting who's in the top 100, accuracy increases considerably as we include the top 200.

```python
count = 0
for artist in artists.keys():
    if artist in overall.keys():
        count += 1

count / len(artists), count / len(overall)
```

This is also really impressive. 88% of all artists that users impute onto Kevin are indeed in his listening record. That sad 7% number is mostly due to the sheer variety of artists that 809 Kevins generate; keep in mind, there are only 201 Kevin Nguyen playlists out there.

That 47% figure is interesting. At what $n$ of the top $n$ artists does similarity peak? We're going to calculate proportion for every single $n$ up to the total size of the playlist sample, since actual Kevin population is ridiculously big. Let's bring in the big guns.

```python
full_range = len(artists)
plst = artists.most_common(full_range)
kn = [artist for artist, num in overall.most_common(full_range)]
```

```python
import numpy as np

ranges = np.arange(1, len(artists))
proportions = np.array([artist_props(plst[:steps], kn[:steps]) for steps in ranges])
proportions.max()
```

So it seems like the similarity peaks at 49.1%, but this EDA is somewhat incomplete without a more thorough anaylsis. Let's plot the progression to get a better idea of what this is doing.

```python
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import warnings; warnings.simplefilter('ignore')
plt.style.use('fivethirtyeight')
import plotly.express as px

df = pd.DataFrame().assign(x=ranges, y=proportions)
fig = px.line(df, x='x', y = 'y', title = 'similarity in artists between perceived and sampled Kevin Nguyens')
fig.show()
```
